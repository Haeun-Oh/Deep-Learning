```python
import tensorflow as tf
tf.__version__
```

```python
# Multi Layer Perceptron

x = [[0, 0], [0, 1], [1, 0], [1, 1]]
y = [[0], [1], [1], [0]]

X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

W1 = tf.Variable(tf.random_normal([2, 3], -1., 1.)) # -1.0 ~ 1.0 사이의 1차원 값
b1 = tf.Variable(tf.random_normal([3], -1., 1.))
L1 = tf.sigmoid(tf.matmul(X, W1)+b1)

W2 = tf.Variable(tf.random_normal([3, 1], -1., 1.)) # -1.0 ~ 1.0 사이의 1차원 값
b2 = tf.Variable(tf.random_normal([1], -1., 1.))
logits = tf.add(tf.matmul(L1, W2), b2)

output = tf.nn.sigmoid(logits)

cost = -tf.reduce_mean(Y*tf.log(output) + (1-Y)*tf.log(1-output))
opt = tf.train.GradientDescentOptimizer(learning_rate=1) # 경사 하강 optimizer
train_op = opt.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    for step in range(8000):
        _, cost_val = sess.run([train_op, cost], feed_dict={X : x, Y : y})
        print(step, cost_val)

    print(sess.run(output, feed_dict={X: x}))
```

```python
# Multi Layer Perceptron

x = [[0, 0], [0, 1], [1, 0], [1, 1]]
y = [[1, 0], [0, 1], [0, 1], [1, 0]] # one hot encoding

X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 2])

W1 = tf.Variable(tf.random_normal([2, 3], -1., 1.)) # -1.0 ~ 1.0 사이의 1차원 값
b1 = tf.Variable(tf.random_normal([3], -1., 1.))
L1 = tf.sigmoid(tf.matmul(X, W1)+b1)

W2 = tf.Variable(tf.random_normal([3, 2], -1., 1.)) # -1.0 ~ 1.0 사이의 1차원 값
b2 = tf.Variable(tf.random_normal([2], -1., 1.))
logits = tf.add(tf.matmul(L1, W2), b2)

output_softmax = tf.nn.softmax(logits)
output_argmax = tf.argmax(logits, 1)

cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=Y, logits=logits))
opt = tf.train.GradientDescentOptimizer(1) # 경사 하강 optimizer
train_op = opt.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    for step in range(8000):
        _, cost_val = sess.run([train_op, cost], feed_dict={X : x, Y : y})
        if step % 100 ==0:
            print(step, cost_val)

    print("predict : ", sess.run(logits, feed_dict={X: x}))
    print("predict with softmax :", sess.run(output_softmax, feed_dict={X: x}))
    print("predict with argmax :", sess.run(output_argmax, feed_dict={X: x}))
```
